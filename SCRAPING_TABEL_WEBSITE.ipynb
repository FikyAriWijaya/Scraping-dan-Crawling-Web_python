```python
from urllib.request import urlopen, Request
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# URL halaman web yang akan diakses
url = 'https://id.wikipedia.org/wiki/Daftar_film_Indonesia_tahun_2024'

# Membuat permintaan HTTP dengan menambahkan header 'User-Agent'
manipulation = Request(url, headers={'User-Agent':'Mozila/5.0'})

# Membuka URL dan mengambil konten HTML Halaman
htmml_text = urlopen(manipulation)

# Memproses HTML 'html.parser' menggugunakan dengan BeautifulSoup untuk mempermudah analisis
soup = BeautifulSoup(htmml_text, 'html.parser')
```

```python
# Mencari tabel pertama dengan class 'wikitable'
tabel = soup.findAll('table',{'class':'wikitable'})[0]

# Mencari tabel pertama dengan class 'wikitable'
baris = tabel.findAll('tr')
```

```python
# Looping melalui setiap baris dalam tabel
for row in baris:
    # Untuk setiap baris, mencari semua elemen header (<th>) atau data (<td>)
    for cell in row.findAll(['th', 'td']):
        # Mengambil teks dari setiap sel tabel
        cell.get_text()
```

```python
# Membuat list kosong untuk menyimpan hasil
hasil = []

# Looping melalui setiap baris dalam tabel
for row in baris:
    # Membuat list kosong untuk menyimpan data tiap baris
    info = []

    # Untuk setiap baris, cari semua sel (baik header <th> atau data <td>)
    for cell in row.findAll(['th', 'td']):
        # Ambil teks dari sel dan hapus spasi di awal/akhir
        info.append(cell.get_text().strip())

    # Tambahkan data baris yang sudah diproses ke dalam hasil
    hasil.append(info)
```

```python
data = pd.DataFrame(hasil)
data
```

```python
# Mengambil baris pertama sebagai header
header = data.iloc[0]

# Menghapus baris pertama yang sudah dijadikan header
data = data[1:]

# Menetapkan baris pertama sebagai nama kolom
data.columns = header

# Mengatur ulang index tanpa inplace
data = data.reset_index(drop=True)
```
